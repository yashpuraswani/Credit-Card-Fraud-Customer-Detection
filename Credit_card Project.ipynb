# Handing Outliers

````python
def find_outliers(data,var):
    q1,q3=data[var].quantile([0.25,0.75])
    IQR=q3-q1
    lower=q1-(1.5*IQR)
    upper=q3+(1.5*IQR)
    return lower,upper
````

````python

lower,upper=find_outliers(num_cols,'RevolvingUtilizationOfUnsecuredLines_yeo')

# Replace lower and upper values in place of outliers
num_cols['RevolvingUtilizationOfUnsecuredLines_yeo']=np.where(num_cols['RevolvingUtilizationOfUnsecuredLines_yeo']<lower,lower,num_cols['RevolvingUtilizationOfUnsecuredLines_yeo'])
num_cols['RevolvingUtilizationOfUnsecuredLines_yeo']=np.where(num_cols['RevolvingUtilizationOfUnsecuredLines_yeo']>upper,upper,num_cols['RevolvingUtilizationOfUnsecuredLines_yeo'])

````

````python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB,BernoulliNB,GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
````

````python
def RF(x_train,y_train,x_test,y_test):
  rf=RandomForestClassifier()
  rf.fit(x_train,y_train)
  print(" Tranning Accuracy : ",rf.score(x_train,y_train))
  print(" Test Accuracy : ",rf.score(x_test,y_test))
  y_predit=rf.predict(x_test)
  print(" Confusion Matrix ",confusion_matrix(y_test,y_predit))
  print(" Classfication Report ",classification_report(y_test,y_predit))
````

````python
def DT(x_train,y_train,x_test,y_test):
  dt=DecisionTreeClassifier()
  dt.fit(x_train,y_train)
  print(" Tranning Accuracy : ",dt.score(x_train,y_train))
  print(" Test Accuracy : ",dt.score(x_test,y_test))
  y_predit=dt.predict(x_test)
  print(" Confusion Matrix ",confusion_matrix(y_test,y_predit))
  print(" Classfication Report ",classification_report(y_test,y_predit))
````

````python
def KNN(x_train,y_train,x_test,y_test):
  knn=KNeighborsClassifier()
  knn.fit(x_train,y_train)
  print(" Tranning Accuracy : ",knn.score(x_train,y_train))
  print(" Test Accuracy : ",knn.score(x_test,y_test))
  y_predit=knn.predict(x_test)
  print(" Confusion Matrix ",confusion_matrix(y_test,y_predit))
  print(" Classfication Report ",classification_report(y_test,y_predit))
````

````python
def logistic_reg(x_train,y_train,x_test,y_test):
  lr=LogisticRegression()
  lr.fit(x_train,y_train)
  print(" Tranning Accuracy : ",lr.score(x_train,y_train))
  print(" Test Accuracy : ",lr.score(x_test,y_test))
  y_predit=lr.predict(x_test)
  print(" Confusion Matrix ",confusion_matrix(y_test,y_predit))
  print(" Classfication Report ",classification_report(y_test,y_predit))
````

````python
def naive_bayes(x_train,y_train,x_test,y_test):
  nb=GaussianNB()
  nb.fit(x_train,y_train)
  print(" Tranning Accuracy : ",nb.score(x_train,y_train))
  print(" Test Accuracy : ",nb.score(x_test,y_test))
  y_predit=nb.predict(x_test)
  print(" Confusion Matrix ",confusion_matrix(y_test,y_predit))
  print(" Classfication Report ",classification_report(y_test,y_predit))
````

````python
def call_all_algo(x_train,y_train,x_test,y_test):
  print(" ------ KNN IMPLEMENTATION ----------")
  KNN(x_train,y_train,x_test,y_test)
  print(" ------ LOGISTIC REGRESSION ----------")
  logistic_reg(x_train,y_train,x_test,y_test)
  print(" ------ NAIVE BAYES----------")
  naive_bayes(x_train,y_train,x_test,y_test)
  print(" ------ DECISION TREE----------")
  DT(x_train,y_train,x_test,y_test)
  print(" ------ RANDOM FOREST----------")
  RF(x_train,y_train,x_test,y_test)
````
